```python
# read data and handle null values

import pandas as pd
df = pd.read_csv("orders.csv", na_values = ["Not Available","unknown"])
df.head(20)

df["Ship Mode"].unique()

df.isnull().sum()
```

```python
# rename column names

# df.rename(columns = {"Order Id":"order_id","City":"city"})

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df.columns
```

```python
# convert order_date to datetime datatype

df.dtypes

df["order_date"] = pd.to_datetime(df["order_date"], format = "%Y-%m-%d")
df.dtypes
```

```python
# derive new columns discount,sale price and profit

df["discount"] = df["list_price"] * df["discount_percent"] /100

df["sale_price"] = df["list_price"] - df["discount"]

df["profit"] = df["sale_price"] - df["cost_price"]

df
```

```python
# drop cost_price, list_price, discount_percent columns

df.drop(columns = ["list_price", "cost_price", "discount_percent"], inplace = True)
df
```

```python
# load the data into sql server

import sqlalchemy as sal

server = "DESKTOP-TSJJHDQ\SQLEXPRESS"
database = "retail_orders"

engine = sal.create_engine(f'mssql+pyodbc://{server}/{database}?trusted_connection=yes&driver=ODBC+Driver+17+for+SQL+Server')
conn = engine.connect()

df.to_sql("orders", con = conn, index = False, if_exists = "replace")
```
